### 4. What percentage of people with advanced education (Bachelors, Masters or Doctorate) make more than 50K?

### 8. What country has the highest percentage of people that earn >50K and what is that percentage?

```python
hs_df = df[df.salary == '>50K']
hs_df.head()
```

```python
salary_table = hs_df['native-country'].value_counts() / df['native-country'].value_counts()
```

```python
salary_table.sort_values(ascending=False)
```

```python
salary_table.max() * 100
```

```python
df[df['native-country'] == 'United-States']['salary'].value_counts(normalize=True)
```

### 9. What is the most popular occupation for those who earn >50K in India?

```python
def age_group(age):
    if 17 <= age <= 32:
        return 'Group 1'
    elif 33 <= age <= 46:
        return 'Group 2'
    elif 47 <= age <= 65:
        return 'Group 3'
    else:
        return 'Group 4'
```

```python
df.age.apply(age_group).value_counts().plot(kind='bar', title='TITLE!')
```
