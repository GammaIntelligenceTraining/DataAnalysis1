### 4. What percentage of people with advanced education (Bachelors, Masters or Doctorate) make more than 50K?

```python
df.education.unique()
```

```python
df.salary.unique()
```

```python
edu_dict = dict(zip(df.education.unique(), df['education-num'].unique()))
edu_dict
```

```python
hedu_df = df[(df['education-num'] >= 13) & (df['education-num'] != 15)]
hedu_df.education.unique()
```

```python
round(hedu_df.salary.value_counts(normalize=True)['>50K'] * 100, 1)
```

### 5. What percentage of people without advanced education make more than 50K?

```python
ledu_df = df[(df['education-num'] < 13) | (df['education-num'] == 15)]
ledu_df.education.unique()
```

```python
round(ledu_df.salary.value_counts(normalize=True)['>50K'] * 100, 1)
```

```python
minh_df.shape
```

```python
minh_df.salary.value_counts(normalize=True)['>50K'] * 100
```

### 8. What country has the highest percentage of people that earn >50K and what is that percentage?

```python
hs_df = df[df.salary == '>50K']
hs_df.head()
```

```python
salary_table = hs_df['native-country'].value_counts() / df['native-country'].value_counts()
```

```python
salary_table.sort_values(ascending=False)
```

```python
salary_table.max() * 100
```
